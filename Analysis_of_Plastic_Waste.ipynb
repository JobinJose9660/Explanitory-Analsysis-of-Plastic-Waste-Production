#Data is cleaned data

Does Total Waste production Increases when production increases?

```python

plt.plot(Pl_Waste_Prod['Total contribution to SUP waste'], Pl_Waste_Prod['Production of in-scope polymers'], c = 'y')
plt.xlabel('Production of Polymers')
plt.ylabel('Total waste')
plt.title('Analysis of Total  Plastic Waste over Production')
plt.grid()
plt.show()
```

We can see that when production of Polymers increase the total waste contribution is increasing
More we increase the  production of plastic polymer it will generate more plastic waste which is harmful to biodiversity


Recommendation:
Decrease the production of polymer ,total waste contribution will be decreased

Does companies with more asset produce more waste ?

```python
sns.scatterplot(x='No. of assets', y = 'Production of in-scope polymers', data = Pl_Waste_Prod)
plt.figure(figsize=(10,8))
plt.show()
```

We can see a positive corelation between 'No.of assets companies owned and production of in - scope polymers, which will result in production of more polymer waste , so if a company have more assets it will produce more waste

Which Polymer format contirbutes more  towards total waste production ?

```python
plt.hist((Pl_Waste_Prod['Flexible format contribution to SUP waste'],Pl_Waste_Prod['Rigid format contribution to SUP waste']),bins=10)
sns.set()
plt.xlabel('Contribution for total waste production')
plt.ylabel('Count')
plt.legend({'Flexible format','Rigid format'})
plt.show()
```

We can see almost both rigid and flexibile contribute equally towards total waste production but flexibile waste production slightly contributes more than rigid format  

Let's check corelation, does ranking of companies related to no. assets they posses ??

```python
#Checking Corelations
corr = Pl_Waste_Prod.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap = 'Greens')
```
